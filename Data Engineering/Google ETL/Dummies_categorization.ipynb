```python
df['total'] = df.sum(axis=1)
df[['gmap_id','total']]
```

```python
def get_category_string(row):
    categories = row.index[1:]  # Excluir la primera columna ('gmap_id')
    selected_categories = [category.split('_')[-1] for category in categories if row[category] == 1]
    return ', '.join(selected_categories)

df['Selected_Categories'] = df.apply(get_category_string, axis=1)
```

```python
df = df[['gmap_id','Selected_Categories']]
df.head()
```

```python
import re

# Definir una función para limpiar el texto
def clean_text(text):
    # Eliminar caracteres especiales, números y puntuación
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Convertir a minúsculas
    text = text.lower()
    return text

# Aplicar la limpieza a la columna 'texto' en tu DataFrame
df['cleaned_text'] = df['Selected_Categories'].apply(clean_text)

df.head()
```

```python
df.head()
```

```python
import pandas as pd

# Supongamos que 'df' es tu DataFrame y 'keyword_lists' es la lista de listas de palabras clave

dictionary = {
    'No Detail':['restaurant'],
    'American': ['american', 'hamburger', 'tex', 'chicken', 'barbecue','barbeque', 'wings', 'steakhouse', 'grill', 'hot','dog'],
    'Central American': ['mexican', 'mex', 'latin', 'salvadoran', 'taco', 'burrito', 'jamaican', 'haitian', 'honduran', 'southern', 'caribbean', 'dominican', 'hawaiian', 'puerto', 'rican', 'cuban', 'guatemalan'],
    'South American': ['latin', 'southern', 'colombian', 'peruvian', 'brazilian', 'venezuelan','argentine','ecuadorian'],
    'Vegetarian': ['vegetarian', 'vegan', 'lunch', 'noodle', 'modern'],
    'Breakfast': ['breakfast', 'brunch', 'house', 'bakery','waffles'],
    'Family': ['family', 'lunch', 'food', 'caterer', 'buffet', 'authentic', 'dining', 'fine', 'traditional', 'sum'],
    'Cafe': ['cafe', 'tea', 'internet', 'house', 'cafeteria', 'art', 'coffee', 'bakery','waffles'],
    'European': ['russian', 'portuguese', 'chips', 'fish', 'bistro', 'italian', 'mediterranean', 'french', 'greek', 'spanish','pizza'],
    'Night': ['soul', 'fusion', 'brewery', 'pub', 'gastropub', 'drink','cajun'],
    'Fastfood': ['takeaway', 'takeout', 'fried', 'chips', 'deli', 'fast'],
    'Asian': ['japanese', 'sushi', 'thai', 'ramen', 'teppanyaki', 'izakaya', 'shabu', 'asian', 'southeast', 'vietnamese', 'chinese', 'korean', 'eastern', 'filipino', 'pakistani', 'indian','uzbeki','romanian'],
    'African': ['african', 'ethiopian', 'ramen', 'halal', 'turkish', 'kosher','arabic'],
    'Seafood':['seafood']
}

df['Predicted_Category'] = None
df['Counts'] = None
# Crea una lista para almacenar las etiquetas de cada restaurante
predicted_categories = []

# Itera sobre cada restaurante
for index, row in df.iterrows():
    # Inicializa un contador para cada categoría
    category_counts = {category: 0 for category in dictionary.keys()}

    # Itera sobre las palabras clave de cada categoría
    for category, keywords in dictionary.items():
        # Incrementa el contador si una palabra clave está presente en el texto
        category_counts[category] += sum(keyword in row['cleaned_text'] for keyword in keywords)

    if not all(value == 0 for key, value in category_counts.items() if key != 'No Detail'):
        del category_counts['No Detail']

    # Obtiene la categoría con la cuenta más alta
    predicted_category = max(category_counts, key=category_counts.get)

    # Agrega la categoría predicha a la lista
    row['Predicted_Category'] = predicted_category
    row['Counts'] = category_counts[predicted_category]

# Muestra el DataFrame resultante con las predicciones de categoría
df.head()
```
